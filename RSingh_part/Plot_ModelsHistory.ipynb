# MODEL 1:

### Overview
A basic Convolutional Neural Network (CNN) designed for multi-class classification.

### Model Layers
1.⁠ ⁠*Convolutional Layers*: Extracts spatial features from the input images.
2.⁠ ⁠*Flattening Layer*: Converts 2D feature maps into a 1D vector for input into the Dense layers.
3.⁠ ⁠*Dense Layers*: Final layers responsible for classification.

### Key Aspects
1. **Epochs**: The model is trained for **20 epochs** with early stopping, allowing the training to halt sooner if the validation loss stops improving. Optimal performance is generally reached at **epoch 8**.
2. **Loss Function**: **Categorical cross-entropy** is used to measure prediction error for multi-class classification, optimizing the model for accuracy across four classes.
3. **Early Stopping**: Early stopping is applied with a patience of 8, meaning training will stop if there is no improvement in validation loss for 8 consecutive epochs. This helps prevent overfitting by halting training when the model reaches its optimal validation performance.
4. **Optimizer**: The **Adam optimizer** is used for efficient training,
5. **Graphs for Loss and Accuracy**:
   - **Training and Validation Accuracy**: Provides a visual of how well the model learns across epochs.
   - **Training and Validation Loss**: Reflects how the model's error rate decreases with training and stabilizes as early stopping is triggered.

This setup helps the model generalize by stopping at its optimal performance.








```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = './Project4_CSVFiles/model1/h1_df.csv'  # Update this to your actual file path if needed
df = pd.read_csv(file_path)

# Plotting the graphs

# 1. Line Plot of Accuracy over Epochs
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['accuracy'], label='Train Accuracy', color='blue')
plt.plot(df.index, df['val_accuracy'], label='Val Accuracy', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Accuracy Over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# 2. Line Plot of Loss over Epochs
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['loss'], label='Training Loss', color='blue')
plt.plot(df.index, df['val_loss'], label='Validation Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

## Optimization Method 2: Adjusting Learning Rate and Implementing Callbacks

In this phase of model optimization, several techniques were employed to further increase accuracy:

### Key Enhancements
- **Adjusting Learning Rate**: The new learning rate is set to **0.0005**, allowing the model to take smaller steps during training. This adjustment can lead to better results as it navigates the loss landscape more precisely.

- **Model Checkpoint**: This callback saves the best-performing model based on monitored metrics (such as validation accuracy). This ensures that the most effective version of the model is preserved throughout training.

- **ReduceLROnPlateau**: This callback monitors the validation accuracy metric and reduces the learning rate when no improvement is observed, allowing the model to make finer adjustments as it approaches optimal performance.

The model was run for a total of **20 epochs** without early stopping. Here, we can see further improvements in accuracy as a result of these optimization techniques.
